```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
text_data = [
    "GFG is providing a new Deep Learning Course which is really good",
    "We will be studying Deep Learning from today",
    "I want a Deep sleep today"
]
```

```python
vec = CountVectorizer(ngram_range=(2,3))
```

```python
vec.fit(text_data)
```

```python
data = vec.transform(text_data).toarray()
```

```python
df = pd.DataFrame(data,columns= vec.get_feature_names_out())
```
