## 2. Preprocessing the DataSet

### 01. Removing special character

```python
for i in '!@#$%^&*()-_=+[]\{}|;:",<.>/?~`\n':
    text_data = text_data.replace(i," ")
text_data = text_data.replace("  "," ")
text_data = text_data.replace("  "," ")
text_data = text_data.replace("   "," ")
text_data = text_data.replace("   "," ")
```

### 02. Lowerize the data

```python
text_data = text_data.lower()
```

```python
text_data[:100]
```

### 03. Tokenizing the data

```python
tokenize_text = word_tokenize(text_data)
```

```python
tokenize_text
```

### 04. Removing the null values

```python
words = [word for word in tokenize_text if word != ' ']
```

### 05. Creating word index

```python
word_index = { word:index for index , word in enumerate(set(words))}
index_of_word = { index : word for index, word in enumerate(set(words))}
```

### 06. Create sentence tokenization

```python
sents = [word_tokenize(sent) for sent in sent_tokenize(text_data)]
```

## 03. Creating Trainable data set

```python
def create_context_target(sent, window_size):
    #extracting features and labels from the sentence
    features = []
    labels =[]
    for sent in sents:
        for i in range(len(sent) - window_size*2):
            features.append(sent[i:i+window_size] + sent[i+window_size+1 : i+window_size*2+1 ])
            labels.append(sent[i+window_size])
    return [features , labels]

```

```python
#create window size in here
WINDOW_SIZE = 2
```

```python
features, labels = create_context_target(sents, WINDOW_SIZE)
```

```python
word_index['15']
```

```python
def OneHotEncode(features, labels):
    #converting features into one hot encoded vectors    
    x_train = []
    for feature in features:
        encode = np.zeros(len(word_index))
        for word in feature:
            encode[int(word_index[word])] = 1
        x_train.append(encode)
    x_train = np.array(x_train)
    
    #converting labels into one hot encoded vectors
    y_train = []
    for label in labels:
        enc = np.zeros(len(word_index))
        enc[word_index[label]] = 1
        y_train.append(enc)
    y_train = np.array(y_train)
    
    return [x_train, y_train]
```

```python
x_train, y_train = OneHotEncode(features, labels)
```
